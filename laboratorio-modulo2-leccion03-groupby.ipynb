#### Ejercicio 5: ¿Cuántos títulos se añadieron a Netflix cada año? Para resolver este ejercicio tendrás uqe convertir la columna `date_added` a formato fecha usando el método `pd.to_datetime`. [Aquí](https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html) tienes la documentación, leeala con detenimiento antes de usar el método. 

```python
df["Fecha_String"] = df["date_added"].astype(str)
df['Fecha_String'] = df['Fecha_String'].str.replace(',', ' ')
df['Estreno'] = pd.to_datetime(df['Fecha_String'], format='mixed')
df['Dia_Semana'] = df['Estreno'].dt.day_name()
df["Año_Estreno"] = df["Estreno"].dt.year.fillna(0).astype(int)
df["Ano"] = df["Ano"].fillna(0).astype(int)

titulos_ano = df.groupby("Año_Estreno")["Title"].count().sort_values(ascending=False)
df_estrenos_ano = titulos_ano.reset_index()

df_estrenos_ano["%"] = np.round(df_estrenos_ano["Title"] / df.shape[0] * 100, 2)
df_estrenos_ano
```

```python
df_top_5_estreno = df_estrenos_ano.nlargest(5, 'Title')
df_top_5_estreno
```

```python
top_5_anos_titulos = df_top_5_estreno["Title"].sum()
total_5_anos_porcentaje = df_top_5_estreno["%"].sum()

print("Total de títulos en los 5 mayores años:", top_5_anos_titulos)
print("Porcentaje total en los 5 mayores años:", total_5_anos_porcentaje)
```

```python
total_anos = len(df["Año_Estreno"].unique())
total_anos
```

- Vemos que que hay un total de 15 años de estrenos. De los cuales 5 (2017-2021) representan el 93% de los estrenos totales

#### Ejercicio 6: ¿Cuál es el país con más títulos lanzados en el año 2020? Para resolver este ejercicio después del gropuby os puede resultar interesante el método [`idxmax`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.idxmax.html)

```python
df_2020 = df[df["Año_Estreno"] == 2020]
titulos_por_pais = df_2020.groupby("country")["Title"].count()

# titulos_por_pais es una Serie de pandas, con lo que los índice de esta serie correponderán a "country" y los valores a el .count() de "Titles"
# Así, usando idxmax() accedemos al índice asociado al valor máximo (es decir, el índice del valor que sacamos usando .max())
pais_con_mas_titulos = titulos_por_pais.idxmax()
titulos_totales = titulos_por_pais.max()

print(pais_con_mas_titulos)
print(titulos_totales)
```
